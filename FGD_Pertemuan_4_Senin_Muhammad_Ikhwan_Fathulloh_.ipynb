6. buatlah matrix (ukurannya bebas) dengan numpy dan hitunglah determinan dan invers dari matrix tersebut


```python
xa = np.array([[1, 5, 3], [1, 2, 9], [2, 1, 1]])
xb = np.array([16, 8, 7])
```

```python
print('det ', np.linalg.det(xa))
print('inv ', np.linalg.inv(xa))
```

7. buatlah 3 persamaan linear dengan 3 variabel ( dengan bentuk umum ax + by + cz = d) dan carilah nilai x,y,z tersebut

```python
x,y,z = np.linalg.solve(xa, xb)
print('hasil x ', x)
print('hasil y ', y)
print('hasil z ', z)
```
